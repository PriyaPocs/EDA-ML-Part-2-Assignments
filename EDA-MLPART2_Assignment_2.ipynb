Decision Making: In some cases, outliers might represent genuine but rare occurrences. The impact of outliers on decision-making depends on the context and goals of the analysis.

Visualization: Outliers can affect data visualization by distorting the scale of plots, making patterns less visible, or creating a cluttered appearance.

It's important to carefully examine any potential outliers you identify using the boxplot. Consider the following steps:

Check the data source: Verify whether the outlier is due to data entry errors, measurement errors, or genuinely extreme values.
Domain knowledge: Consult domain experts to determine if the outlier is plausible within the context of the problem.
Data transformation: Depending on the analysis, you might choose to remove, transform, or treat outliers in a specific way.

Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset to identify the peak travel season. What features would you analyze to identify the peak season, and how would you present your findings to your boss?

Answer(Q6):


```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar

# Load the flight price dataset
df=pd.read_excel('flight_price.xlsx')
df = df.dropna()
no_duplicate=df.drop_duplicates(keep='first')
flight_data = no_duplicate.copy()

# Convert 'Date' column to datetime
flight_data['Date'] = pd.to_datetime(flight_data['Date_of_Journey'])

# Extract month, day of the week, and year from the 'Date' column
flight_data['Month'] = flight_data['Date'].dt.month
flight_data['DayOfWeek'] = flight_data['Date'].dt.dayofweek
flight_data['Year'] = flight_data['Date'].dt.year

# Analyze monthly average flight prices
monthly_avg_prices = flight_data.groupby('Month')['Price'].mean()

# Analyze day of the week average flight prices
day_of_week_avg_prices = flight_data.groupby('DayOfWeek')['Price'].mean()

# Analyze average flight prices by destination
destination_avg_prices = flight_data.groupby('Destination')['Price'].mean()

# Create subplots to display the analysis results
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Find the peak travel season (month with the highest average price)
peak_month = monthly_avg_prices.idxmax()
# print("peak_month = ",peak_month)
# Get the month name from the month number
peak_month_name = calendar.month_name[peak_month]

# Monthly Analysis
axes[0, 0].plot(monthly_avg_prices, marker='o')
axes[0, 0].set_xticks(range(1, 13))
axes[0, 0].set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
axes[0, 0].set_title('Monthly Average Flight Prices')
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Average Price')
axes[0, 0].annotate(f'Peak Travel Season(Month): {peak_month_name}', xy=(peak_month, monthly_avg_prices.max()), xytext=(peak_month + 0.5, monthly_avg_prices.max() + 50),
             arrowprops=dict(arrowstyle='->',color='r'), fontsize=10,c='b')

# Day of the Week Analysis
peak_day = day_of_week_avg_prices.idxmax()
# print("peak_day = ",peak_day)
# Get the month name from the month number
peak_day_name = calendar.day_name[peak_day]
# print("peak_day_name = ",peak_day_name)
day_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
axes[0, 1].plot(day_of_week_avg_prices, marker='o')
axes[0, 1].set_xticks(range(7))
axes[0, 1].set_xticklabels(day_names)
axes[0, 1].set_title('Day of the Week Average Flight Prices')
axes[0, 1].set_xlabel('Day of the Week')
axes[0, 1].set_ylabel('Average Price')
axes[0, 1].annotate(f'Peak Travel Season(Day): {peak_day_name}', xy=(peak_day, day_of_week_avg_prices.max()), xytext=(peak_day + 0.5, day_of_week_avg_prices.max() + 50),
             arrowprops=dict(arrowstyle='->',color='r'), fontsize=10,c='b')

# Destination-specific Analysis (showing top destinations)
top_destinations = destination_avg_prices.sort_values(ascending=False).head(5)
top_destinations.plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Top Destination Average Flight Prices')
axes[1, 0].set_xlabel('Destination')
axes[1, 0].set_ylabel('Average Price')

# Adjust layout and display the subplots
plt.tight_layout()
plt.show()
```

1. Peak travel season identified in the month of January.

```python
# Function to convert duration to minutes
def convert_duration_to_hrs(duration_str):
    hours, minutes = 0, 0
    parts = duration_str.split()

    for part in parts:
        if 'h' in part:
            hours = int(part.replace('h', ''))
        elif 'm' in part:
            minutes = int(part.replace('m', ''))
    total_hrs = hours + minutes/60
    return round(total_hrs)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the flight price dataset 
df=pd.read_excel('flight_price.xlsx')
no_duplicate=df.drop_duplicates(keep='first')
flight_data = no_duplicate.copy()

# Convert 'Date' column to datetime
flight_data['Date'] = pd.to_datetime(flight_data['Date_of_Journey'])

# Create subplots for visualizations
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Time Series Plot: Monthly Trends
monthly_avg_prices = flight_data.groupby(flight_data['Date'].dt.to_period('M'))['Price'].mean()
monthly_avg_prices.plot(ax=axes[0, 0])
axes[0, 0].set_title('Monthly Average Flight Prices')
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Average Price')

# Box Plot: Day of the Week Trends
flight_data['DayOfWeek'] = flight_data['Date'].dt.strftime('%A')
flight_data['DayOfWeek']
daily_avg_prices = flight_data.groupby(flight_data['DayOfWeek'])['Price'].mean()
daily_avg_prices.plot(kind='bar',ax=axes[0, 1])
axes[0, 1].set_title('Flight Prices by Day of the Week')
axes[0, 1].set_xlabel('Day of the Week')
axes[0, 1].set_ylabel('Price')

# Bar Plot: Destination Analysis
top_destinations = flight_data['Destination'].value_counts().head(5)
top_destinations.plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Top Destinations')
axes[1, 0].set_xlabel('Destination')
axes[1, 0].set_ylabel('Number of Flights')

# Bar graph for Total_Stops vs Price
avg_price_stops = df.groupby('Total_Stops')['Price'].mean().reset_index()
avg_price_stops.plot(kind='bar', ax=axes[1, 1])
axes[1, 1].set_title('Total Stops Vs Average Price')
axes[1, 1].set_xlabel('Total Stops')
axes[1, 1].set_ylabel('Average Price (in currency)')

# Adjust layout and display the subplots
plt.tight_layout()
plt.show()
```

Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

Answer(Q12):

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Play Store dataset (replace 'googleplaystore.csv' with your dataset file)
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
df_copy=df.drop_duplicates(subset=['App'],keep='first')

# Remove rows with missing 'Size' or 'Rating' values
df_cleaned = df.dropna(subset=['Size', 'Rating'])


# Convert 'Size' to a numerical value (e.g., in MB)
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024
    else:
        return None

df_cleaned['Size'] = df_cleaned['Size'].apply(convert_size)
df_cleaned['Size'] = df_cleaned['Size'].astype(float)
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned['Size'], df_cleaned['Rating'], alpha=0.5)
plt.title('Relationship between App Size and Rating')
plt.ylabel('Size (MB)')
plt.xlabel('Rating')
plt.grid(True)
plt.show()
```

Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?


Answer(Q16):


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the Google Play Store dataset 
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
# print(df)
df_copy=df.drop_duplicates(subset=['App'],keep='first')
df_copy=df_copy.dropna()

df_copy['Price'] = df_copy['Price'].str.replace('$', '')
df_copy['Price'] = df_copy['Price'].astype(float)

df_copy['Installs'] = df_copy['Installs'].str.replace('+', '')
df_copy['Installs'] = df_copy['Installs'].str.replace(',', '')
df_copy['Installs'] = df_copy['Installs'].astype(int)
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
df_copy.columns
data = df_copy.copy()

# Function to extract developer information from app names
def extract_developer(app_name):
    parts = app_name.split(' - ')
    if len(parts) > 1:
        return parts[0]
    # If no hyphen pattern found, return the entire app name
    return app_name

# Apply the function to create a 'Developer' column
data['Developer'] = data['App'].apply(extract_developer)
# Number of Apps by Developer
apps_per_developer = data['App'].groupby(data['Developer']).count()

# Total Installs by Developer
installs_per_developer = data['Installs'].groupby(data['Developer']).sum()

# Average Rating by Developer
avg_rating_per_developer = data['Rating'].groupby(data['Developer']).mean()

# Total Reviews by Developer
total_reviews_per_developer = data['Reviews'].groupby(data['Developer']).sum()

# Visualization
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
apps_per_developer.nlargest(10).plot(kind='bar', title='Number of Apps by Developer')

plt.subplot(2, 2, 2)
installs_per_developer.nlargest(10).plot(kind='bar', title='Total Installs by Developer')

plt.subplot(2, 2, 3)
avg_rating_per_developer.nlargest(10).plot(kind='bar', title='Average Rating by Developer')

plt.subplot(2, 2, 4)
total_reviews_per_developer.nlargest(10).plot(kind='bar', title='Total Reviews by Developer')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the Google Play Store dataset 
df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
# print(df)
df_copy=df.drop_duplicates(subset=['App'],keep='first')
df_copy=df_copy.dropna()

df_copy['Price'] = df_copy['Price'].str.replace('$', '')
df_copy['Price'] = df_copy['Price'].astype(float)

df_copy['Installs'] = df_copy['Installs'].str.replace('+', '')
df_copy['Installs'] = df_copy['Installs'].str.replace(',', '')
df_copy['Installs'] = df_copy['Installs'].astype(int)
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
df_copy.columns
data = df_copy.copy()

# Convert 'Last Updated' to datetime format
data['Last Updated'] = pd.to_datetime(data['Last Updated'])

# Extract Year and Month from 'Last Updated'
data['Year'] = data['Last Updated'].dt.year
data['Month'] = data['Last Updated'].dt.month

# App Category Analysis
category_installs = data.groupby('Category')['Installs'].sum()

# Installs Analysis Over Time
installs_over_time = data.groupby(['Year', 'Month'])['Installs'].sum()

# Reviews and Ratings Analysis
avg_ratings = data.groupby(['Year', 'Month'])['Rating'].mean()
avg_reviews = data.groupby(['Year', 'Month'])['Reviews'].sum()

# Visualization
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
sns.barplot(x=category_installs.index, y=category_installs.values)
plt.title('Category vs. Total Installs')
plt.xticks(rotation=90)

plt.subplot(2, 2, 2)
installs_over_time.plot(kind='line', marker='o')
plt.title('Installs Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Total Installs')
plt.xticks(rotation=90)

plt.subplot(2, 2, 3)
sns.scatterplot(data=data, x='Reviews', y='Rating')
plt.title('Reviews vs. Ratings')
plt.xticks(rotation=90)

plt.subplot(2, 2, 4)
sns.heatmap(data[['Year', 'Month', 'Installs']].pivot_table(index='Year', columns='Month', values='Installs'), cmap='YlGnBu')
plt.title('Heatmap of Installs by Year-Month')
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()
```
